```python
year = df['year']
month = df['month']
day = df['day']
lat = df['lat']
lon = df['lon']
```

```python
working_dir = 'C:/GGGG/hysplit/working'
output_dir = 'D:/run2/lag+1/0-40'
meteo_dir = 'D:/ncepncar'

for i in range(0,400000):
    pysplit.generate_bulktraj(basename =(str(i))+'_'+str(lat[i])+str(lon[i]), 
                              hysplit_working = working_dir,
                              output_dir = output_dir, 
                              meteo_dir = meteo_dir,
                              years = [year[i]],
                              months = [month[i]],
                              hours = [0],  
                              altitudes = [3000], 
                              coordinates = (lat[i], lon[i]), 
                              run = -216,
                              monthslice = slice(day[i]-1,(day[i])),
                              meteo_bookends=([1], [1]),
                              get_clipped=False, 
                              get_reverse=False,
                              meteoyr_2digits=False,
                              outputyr_2digits=False,
                              hysplit = 'C:/GGGG/hysplit/exec/hyts_std')
```
